# Листья
В исходных данных содержится результат исследования 40 различных сортов листьев и нескольких вариантов для каждого из них.
Набор данных представлен в виде csv файла, со значением характеристик для каждого из образцов. Данные получены на основе 24-bit RGB фотографий образцов, сделанных на Apple IPAD 2 в разрешении 720x920px.

Атрибуты:
1. Class (Species) - Класс (вид) листка
2. Specimen Number - Номер его в наборе для каждого класса
3. Eccentricity - Эксцентриситет
4. Aspect Ratio - Соотношение сторон 
5. Elongation - Удлинение
6. Solidity - Сопротивление
7. Stochastic Convexity - Стохастическая выпуклость
8. Isoperimetric Factor - Изопериметрический коэффициент
9. Maximal Indentation Depth - Максимальная глубина отступа 
10. Lobedness - Гибкость
11. Average Intensity - Средняя интенсивность
12. Average Contrast - Средняя контрастность 
13. Smoothness - Гладкость 
14. Third moment - Третий момент
15. Uniformity -  Однородность
16. Entropy - Энтропия

Для получения подробного описания каждой характеристики можно обратиться здесь https://github.com/DmitryNakhimovich/ML_homework_1/blob/master/ReadMe.pdf

# Подготовка данных

возможно удаление излишних данных, а именно номера в наборе для класса ('Specimen_Number')

```python
data = data.drop(['Specimen_Number'], axis = 1)
```

```python
# выделим количественные и категориальные признаки
numerical_columns = [c for c in data.columns if data[c].dtype.name == 'float64']
print(numerical_columns) 
categorical_columns = [c for c in data.columns if data[c].dtype.name == 'int64']
print(categorical_columns)  
```

```python
# посмотрим уникальные значения категориального признака
for c in categorical_columns:
    print(c, data[c].unique())
```

```python
# проверим на целостность ячейки таблицы
data.count(axis = 0)
```

# Векторизация
у нас только один категориальный признак

```python
nonbinary_columns = categorical_columns
print(nonbinary_columns)
```

```python
# убедимся в небинарности признака
data['Class'].unique()
```

```python
# векторизуем небинарный признак
data_nonbinary = pd.get_dummies(data[nonbinary_columns])
print(data_nonbinary.columns)
```

# Нормализация количественных признаков
Для уменьшения сложности вычислений при масштабировании данных

```python
# выведем количественные признаки
data_numerical = data[numerical_columns]
data_numerical.describe()
```

```python
# нормализуем количественные признаки
data_numerical = (data_numerical - data_numerical.mean(axis = 0))/data_numerical.std(axis = 0)
data_numerical.describe()
```

# Объединение в одну таблицу

```python
data = pd.concat((data_numerical, data_nonbinary), axis = 1)
print(data.shape)
data.describe()
```

# X и y

```python
# Исключаем стоблец 'Class' из набора исходных данных
X = data.drop(('Class'), axis = 1)
# Заносим для корректировки обучения
y = data['Class']
feature_names = X.columns
```

```python
# зададим обучающие и тестовые выборки
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 50)

N_train, _ = X_train.shape 
N_test,  _ = X_test.shape 

print(N_train, N_test)
```

```python
N_train + N_test
```

# kNN
выберем параметр k = 3 и проведем обучение выборке X_train, y_train

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 3) # передаем выбранный параметр k = 3
knn.fit(X_train, y_train)
```

```python
# применяем метод predict, считаем среднюю ошибку
y_train_predict = knn.predict(X_train)
y_test_predict = knn.predict(X_test)

err_train = np.mean(y_train != y_train_predict)
err_test  = np.mean(y_test  != y_test_predict)

print(err_train, err_test)
```

```python
# другой способ посчитать ошибку через knn.score
err_test = 1 - knn.score(X_test, y_test)
err_test
```

c помощью метода confusion_matrix вычислим матрицу, (i,j) элемент которой равен числу объектов из группы i, но по предсказанию находящихся в группе j

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test, y_test_predict))
```

# Подбор параметров
Найдем оптимальное значение параметра k

```python
from sklearn.model_selection import GridSearchCV
nnb = [1, 3, 5, 10, 15, 20, 25, 35, 45, 55]
knn = KNeighborsClassifier()
grid = GridSearchCV(knn, param_grid = {'n_neighbors': nnb}, cv=4)
grid.fit(X_train, y_train)

best_cv_err = 1 - grid.best_score_
best_n_neighbors = grid.best_estimator_.n_neighbors

print(best_cv_err, best_n_neighbors)
```

применим новое значение параметра k = 1

```python
knn = KNeighborsClassifier(n_neighbors = best_n_neighbors).fit(X_train, y_train)

err_train = np.mean(y_train != knn.predict(X_train))
err_test  = np.mean(y_test  != knn.predict(X_test))

print(err_train, err_test)
```

# Домашняя работа №2

На исходных данных проведем задачу классификации. Определим к какому классу (виду) принадлежит листок.
Разобьем данные на обучающую (77%) и тестовую (33%) выборки.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 7)
N_test, _ = X_test.shape
```

Обучаем MLP классификатор.

```python
from sklearn.neural_network import MLPClassifier
mlp_model = MLPClassifier(hidden_layer_sizes = (10), solver = 'lbfgs', activation = 'logistic', random_state = 5)
mlp_model.fit(X_train, y_train)
```

hidden_layer_sizes = 10 - один скрытый слой с 10 нейронами, 

activation = 'logistic' - функция активации для скрытого слоя f(x) = 1 / (1 + exp(-x)).

```python
# Проверяем качество классификатора по величине ошибки:
y_train_pred = mlp_model.predict(X_train)
y_test_pred = mlp_model.predict(X_test)
print("Ошибка на обучающей выборке: ", np.mean(y_train != y_train_pred))
print("Ошибка на тестовой выборке: ", np.mean(y_test != y_test_pred))
```

Увеличим число нейронов на скрытом слое до 100, т.к. ошибка на обучающей выборке меньше, чем на тестовой. Это соответствует ситуации переобучения.

```python
mlp_model = MLPClassifier(hidden_layer_sizes = (100), solver = 'lbfgs', activation = 'logistic', random_state = 5)
mlp_model.fit(X_train, y_train)
```

```python
y_train_pred = mlp_model.predict(X_train)
y_test_pred = mlp_model.predict(X_test)
print("Ошибка на обучающей выборке: ", np.mean(y_train != y_train_pred))
print("Ошибка на тестовой выборке: ", np.mean(y_test != y_test_pred))
```

Величина ошибки усеньшилась. Попробуем увеличить число скрытых слоев до двух, с 10 нейронами в каждой.

```python
mlp_model = MLPClassifier(hidden_layer_sizes = (10, 10), solver = 'lbfgs', activation = 'logistic', random_state = 5)
mlp_model.fit(X_train, y_train)
```

```python
y_train_pred = mlp_model.predict(X_train)
y_test_pred = mlp_model.predict(X_test)
print("Ошибка на обучающей выборке: ", np.mean(y_train != y_train_pred))
print("Ошибка на тестовой выборке: ", np.mean(y_test != y_test_pred))
```

Видим, что величина ошибки резко увеличилась. Следовательно далее используем конфигурацию с одним скрытым слоем и 100 нейронами.

# Подбор параметра alpha

```python
alpha_arr = np.logspace(-5, 1.3, 30)
test_err = []
train_err = []
for alpha in alpha_arr:
    mlp_model = MLPClassifier(alpha = alpha, hidden_layer_sizes = (10), 
                              solver = 'lbfgs', activation = 'logistic', random_state = 5)
    mlp_model.fit(X_train, y_train)

    y_train_pred = mlp_model.predict(X_train)
    y_test_pred = mlp_model.predict(X_test)
    train_err.append(np.mean(y_train != y_train_pred))
    test_err.append(np.mean(y_test != y_test_pred))
```

```python
# Построим график величины ошибки от значения alpha для наглядности
plt.figure(figsize = (16, 6))
plt.semilogx(alpha_arr, train_err, 'b-o', label = 'test')
plt.semilogx(alpha_arr, test_err, 'r-o', label = 'train')
plt.xlim([np.max(alpha_arr), np.min(alpha_arr)])
plt.title('Error vs. alpha')
plt.xlabel('alpha')
plt.ylabel('error')
plt.legend()
pass
```

```python
print(alpha_arr)
```

```python
# Найдем минимальное значение ошибки
min_train_err = np.min(train_err)
min_test_err = np.min(test_err)
print("Минимальная ошибка на обучающей выборке: ", min_train_err)
print("Минимальная ошибка на тестовой выборке: ", min_test_err)
```

```python
# Значения alpha при минимальной величине ошибки
alpha_opt = alpha_arr[test_err == min_test_err]
print(alpha_opt)
```

```python
# Оптимальное значение для alpha
alpha_opt = alpha_opt[0]
print(alpha_opt)
```

## Классификатор при найденном оптимальном значении alpha:

```python
mlp_model = MLPClassifier(alpha = alpha_opt, hidden_layer_sizes = (10),
                          solver = 'lbfgs', activation = 'logistic', random_state = 5)
mlp_model.fit(X_train, y_train)

y_train_pred = mlp_model.predict(X_train)
y_test_pred = mlp_model.predict(X_test)
print("Ошибка на обучающей выборке: ", np.mean(y_train != y_train_pred))
print("Ошибка на тестовой выборке: ", np.mean(y_test != y_test_pred))
```

```python
# Найдем матрицу рассогласования
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test_pred, y_test))
```

```python
M = confusion_matrix(y_test_pred, y_test)
M = np.sqrt(M)
plt.figure(figsize = (10, 10))
plt.imshow(M, interpolation = 'nearest')
plt.set_cmap('binary')
plt.xticks(range(10))
plt.yticks(range(10))
plt.xlabel("true label")
plt.ylabel("predicted label")
plt.colorbar()
pass
```

# Найдём оптимальное число нейронов

```python
neurons_arr = np.arange(2, 20, 1)
test_err = []
train_err = []
for neurons in neurons_arr:
    mlp_model = MLPClassifier(hidden_layer_sizes = (neurons, neurons), 
                              solver = 'lbfgs', activation = 'logistic', random_state = 5)
    mlp_model.fit(X_train, y_train)

    y_train_pred = mlp_model.predict(X_train)
    y_test_pred = mlp_model.predict(X_test)
    train_err.append(np.mean(y_train != y_train_pred))
    test_err.append(np.mean(y_test != y_test_pred))
```

```python
# Построим график зависимости величины ошибки от количества нейронов
plt.figure(figsize = (16, 6))
plt.plot(neurons_arr, train_err, 'b-o', label = 'test')
plt.plot(neurons_arr, test_err, 'r-o', label = 'train')
plt.xlim([np.max(neurons_arr), np.min(neurons_arr)])
plt.title('Error vs. neurons count')
plt.xlabel('neurons count')
plt.ylabel('error')
plt.legend()
pass
```

```python
min_train_err = np.min(train_err)
min_test_err = np.min(test_err)
print("Минимальная ошибка на обучающей выборке: ", min_train_err)
print("Минимальная ошибка на тестовой выборке: ", min_test_err)
```

Оптимальное число нейронов:

```python
neurons_opt = neurons_arr[test_err == min_test_err]
print(neurons_opt)
neurons_opt = neurons_opt[0]
```

# Домашняя работа №3
Будем предсказывать класс (вид) листка по набору его характеристик.

```python
# Разобъем данные на тестовую и обучающую выборки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 6)
```

## Random forest
C помощью случайного леса

```python
# Проведем обучение Random forest классификатора на тренировочной выборке с параметром
from sklearn import ensemble
rf = ensemble.RandomForestClassifier(n_estimators = 50) # n_estimators = 50 (кол-во построенных деревьев)
rf.fit(X_train, y_train) 
```

```python
# Проверим качество классификатора по величине ошибки на тестовой и обучающей выборках
err_train = np.mean(y_train != rf.predict(X_train))
err_test  = np.mean(y_test  != rf.predict(X_test))

print("Ошибка на обучающей выборке: ", err_train)
print("Ошибка на тестовой выборке: ", err_test)
```

Наблюдаем переобучение, т.к. ошибка на тестовой выборке больше, чем на обучающей.

Построим график зависимости ошибки на тестовой и обучающей выборках, а также oob-ошибки от количества деревьев.

```python
# Вычислим значения ошибок для различных деревьев
trees_arr = np.arange(20, 100, 5)
test_err = []
train_err = []
oob_err = []
for trees in trees_arr:
    rf = ensemble.RandomForestClassifier(n_estimators = trees, oob_score=True)
    rf.fit(X_train, y_train)
    train_err.append(np.mean(y_train != rf.predict(X_train)))
    test_err.append(np.mean(y_test  != rf.predict(X_test)))
    oob_err.append(1 - rf.oob_score_)
# Построим график величины ошибки от количества деревьев
plt.figure(figsize = (16, 6))
plt.plot(trees_arr, train_err, 'b-o', label = 'train')
plt.plot(trees_arr, test_err, 'r-o', label = 'test')
plt.plot(trees_arr, oob_err, 'g-o', label = 'oob')
plt.xlim([np.max(trees_arr), np.min(trees_arr)])
plt.title('Error vs. number of trees')
plt.xlabel('trees')
plt.ylabel('error')
plt.legend()
pass
```

Переобучение наблюдается на всех рассматриваемых значениях количества деревьев. Минимум oob-ошибки наблюдается при количестве деревьев 65, 85.

```python
# Минимальное значение ошибки:
min_train_err = np.min(train_err)
min_test_err = np.min(test_err)

print("Минимальная ошибка на обучающей выборке: ", min_train_err)
print("Минимальная ошибка на тестовой выборке: ", min_test_err)
```

```python
# Оптимальное количество деревьев:
trees_opt = trees_arr[test_err == min_test_err]
print(trees_opt)
```

Как видим, размерность данных уменьшилась с 14 до 12.
Далее проведем сравнение ошибок, получаемых при использовании классификатора Extremely Randomized Trees на исходных данных и данных при применении метода главных компонент. 

```python
# Обучаем классификатор вида Extremely Randomized Trees
ert = ensemble.ExtraTreesClassifier(n_estimators = 50) # n_estimators = 50 (кол-во построенных деревьев)
ert.fit(X_train_reduce, y_train)

# Проверим качество классификатора по величине ошибки на тестовой и обучающей выборках
err_train = np.mean(y_train != ert.predict(X_train_reduce))
err_test  = np.mean(y_test  != ert.predict(X_test_reduce))

print("Ошибка на обучающей выборке: ", err_train)
print("Ошибка на тестовой выборке: ", err_test)
```

Наблюдаем переобучение, т.к. ошибка на тестовой выборке больше, чем на обучающей.

После использования PCA ошибки на обчающей выборке совпадают. На тестовой выборке до использования метода главных компонент ошибки имели следующие значения:
Ошибка на обучающей выборке:  0.0;
Ошибка на тестовой выборке:  0.0882352941176;
После применения PCA ошибка на тестовой выборке уменьшилась. Ситуация переобучения так же наблюдается на обоих вариантах.

Построим график зависимости ошибки на тестовой и обучающей выборках и oob-ошибки от количества компонент в методе.

```python
# Вычислим значения ошибок для различного количества компонент
components_arr = np.arange(1, 14, 1)
test_err = []
train_err = []
oob_err = []
for components in components_arr:
    pca = PCA(n_components = components)
    X_train_reduce = pca.fit_transform(X_train)
    X_test_reduce = pca.transform(X_test)
    ert = ensemble.ExtraTreesClassifier(n_estimators = 50, oob_score=True, bootstrap=True) # n_estimators = 50 (кол-во построенных деревьев)
    ert.fit(X_train_reduce, y_train)
    train_err.append(np.mean(y_train != ert.predict(X_train_reduce)))
    test_err.append(np.mean(y_test  != ert.predict(X_test_reduce)))
    oob_err.append(1 - ert.oob_score_)
```

```python
# Построим график величины ошибок
plt.figure(figsize = (16, 6))
plt.plot(components_arr, train_err, 'b-o', label = 'train')
plt.plot(components_arr, test_err, 'r-o', label = 'test')
plt.plot(components_arr, oob_err, 'g-o', label = 'oob')
plt.xlim([np.max(components_arr), np.min(components_arr)])
plt.title('Error vs. number of components')
plt.xlabel('components')
plt.ylabel('error')
plt.legend()
pass
```

Из графика видно, что оптимальными значениями количества компонент являются 12-8. При дальнейшей редукции, величина ошибки начинает увеличиваться. Величина oob-ошибки немного больше чем тестовой выборки и имеет с ней схожий характер роста при уменьшении числа компонент метода.

```python
# Минимальное значение ошибки:
min_train_err = np.min(train_err)
min_test_err = np.min(test_err)

print("Минимальная ошибка на обучающей выборке: ", min_train_err)
print("Минимальная ошибка на тестовой выборке: ", min_test_err)
```

```python
# Оптимальное количество компонент
components_opt = components_arr[test_err == min_test_err]
components_opt = components_opt[0]
print(components_opt)
```
